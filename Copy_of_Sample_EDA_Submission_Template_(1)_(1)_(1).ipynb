````python
# Write a code using .dropna() function using imdb_score column in given dataset

# Count titles before dropping rows with missing IMDb scores
rows_before = titles_df.shape[0]


# Use .dropna() to remove any rows where the 'imdb_score' column is missing (NaN)
titles_df_complete_scores = titles_df.dropna(subset=['imdb_score'])

# Count titles after dropping
rows_after = titles_df_complete_scores.shape[0]
rows_dropped = rows_before - rows_after

print("--- DropNA Example: Removing Missing Scores ---")
print(f"Total Titles (Before dropna): {rows_before}")
print(f"Titles with Complete Scores (After dropna): {rows_after}")
print(f"Number of rows dropped: {rows_dropped}")
````

````python
# write a code using merge  function in given dataset
import pandas as pd

# 1. Load the two DataFrames
# 'titles.csv.zip/titles.csv' contains the title metadata
titles_df = pd.read_csv('titles.csv.zip')
# 'credits.csv.zip/credits.csv' contains the cast and crew information
credits_df = pd.read_csv('credits.csv.zip')

print("Titles DataFrame Head:")
print(titles_df[['id', 'title', 'type']].head())
print("\nCredits DataFrame Head:")
print(credits_df[['id', 'name', 'role']].head())

# 2. Merge the two DataFrames
# We use an 'inner' merge on the common column 'id'. This keeps only rows
# where the 'id' exists in both DataFrames.
merged_df = pd.merge(
    titles_df,
    credits_df,
    on='id',
    how='inner'
)

print("\nMerged DataFrame Head (Titles + Credits):")
print(merged_df[['id', 'title', 'name', 'role']].head())

# 3. Data Cleaning / Filtering Example: Find all Directors
# We filter the merged DataFrame to only include rows where the 'role' is 'DIRECTOR'.
directors_df = merged_df[merged_df['role'] == 'DIRECTOR'].copy()

# Select and display the final cleaned columns
final_directors_list = directors_df[['title', 'name', 'release_year']].sort_values(by='release_year')

print("\nResult of Merging and Cleaning (First 10 Directors):")
print(final_directors_list.head(10))
````

#### Chart - 3 : **Box plot by using imdb_score (Bivariate Analysis)**

````python
# Chart - 3 visualization code
import matplotlib.pyplot as plt
import pandas as pd

# 1. Prepare Data: Filter the titles_df to only include records with an IMDb score
# and select the type and score columns.
plot_data = titles_df[['type', 'imdb_score']].dropna(subset=['imdb_score'])

# Separate the scores into two lists for the box plot function
movie_scores = plot_data[plot_data['type'] == 'MOVIE']['imdb_score']
show_scores = plot_data[plot_data['type'] == 'SHOW']['imdb_score']

data_to_plot = [movie_scores, show_scores]
labels = ['MOVIE', 'SHOW']

# 2. Create the box plot
plt.figure(figsize=(6, 4))

# Plot the data
plt.boxplot(data_to_plot, labels=labels, patch_artist=True,
            boxprops=dict(facecolor='#D3D3D3', color='black'),
            medianprops=dict(color='red', linewidth=2))

# Set labels and title
plt.title('IMDb Score Distribution: Movies vs. TV Shows', pad=20)
plt.ylabel('IMDb Score')
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Save the figure
plt.savefig('imdb_score_boxplot.png')

````

##### 1. Why did you pick the specific chart?

A box plot is a powerful visualization for displaying the distribution, spread, and skewness of a numerical dataset, often used for comparing distributions across different groups.

This box plot is a classic example of Bivariate Analysis as it contains categorical (type) and numerical (imdb_score) for analysis.

So I used a Box plot here to represent coding example that compares the IMDb Score distribution between Movies and TV Shows.

##### 2. What is/are the insight(s) found from the chart?

Insights found from the above chart are as follows:
1. **TV Shows are Consistently Better**

* Higher Middle Score (Median): The middle line (median) in the SHOW box is much higher on the score scale than the middle line in the MOVIE box.

* Higher Overall Distribution: The entire box for the SHOWS is positioned higher than the entire box for the MOVIES.

2. **Movies Have a Wider Range of Quality**
* Wider Box and Whiskers: The box and whiskers for the MOVIE distribution are much longer than those for the TV Shows.

3. **The High-Score Problem**

* Similar Top Outliers: Both the Movie and the Show plot have several dots (outliers) at the very top of the scale (scores 9.0 and above).

#### Chart - 11: **Bubble Chart using Multivariate Analysis**

````python
# Chart - 11 visualization code
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load both datasets
titles_df = pd.read_csv('titles.csv.zip')
credits_df = pd.read_csv('credits.csv.zip')

# 1. Filter for Directors and Count Titles
directors_df = credits_df[credits_df['role'] == 'DIRECTOR'].copy()

# Count the number of titles directed by each person
director_counts = directors_df['name'].value_counts()
TOP_N = 10
top_10_directors = director_counts.head(TOP_N).index.tolist()

# 2. Merge data to get scores and runtimes for the top 10
# Filter directors_df to include only the top 10
top_directors_df = directors_df[directors_df['name'].isin(top_10_directors)].copy()

# Merge with titles_df
merged_df = pd.merge(top_directors_df, titles_df, on='id', how='left')

# 3. Aggregate Data for the Bubble Chart (Trivariate Analysis)
# Calculate the average score and average runtime, and include the title count
director_stats = merged_df.dropna(subset=['imdb_score', 'runtime']).groupby('name').agg(
    avg_score=('imdb_score', 'mean'),
    avg_runtime=('runtime', 'mean'),
    title_count=('id', 'count')
).reset_index()

# Filter again to ensure only the top 10 are used (in case of ties/missing data issues)
director_stats = director_stats[director_stats['name'].isin(top_10_directors)]

# Prepare bubble size (scale the counts for visualization, e.g., square root and multiply)
director_stats['bubble_size'] = np.sqrt(director_stats['title_count']) * 50

# Display the data used for the chart
print("Top 10 Directors: Score, Runtime, and Title Count:")
print(director_stats.sort_values(by='title_count', ascending=False).to_string(index=False))

# 4. Create the Bubble Chart
plt.figure(figsize=(8,5))
scatter = plt.scatter(
    x=director_stats['avg_runtime'],
    y=director_stats['avg_score'],
    s=director_stats['bubble_size'], # Bubble size determined by title count
    alpha=0.6,
    c=plt.cm.viridis(np.linspace(0, 1, len(director_stats))), # Color for differentiation
    edgecolors='k',
    linewidths=0.5
)

# Annotate each bubble with the director's name (only showing top 5 for clarity, rest in legend)
for i in range(len(director_stats)):
    if director_stats['title_count'].iloc[i] >= director_stats['title_count'].nlargest(5).iloc[-1]: # Only label the largest 5
        plt.annotate(
            director_stats['name'].iloc[i],
            (director_stats['avg_runtime'].iloc[i], director_stats['avg_score'].iloc[i]),
            textcoords="offset points",
            xytext=(0, 10),
            ha='center',
            fontsize=9
        )

plt.title('Director Performance: Average Score vs. Average Runtime (Bubble Size = Title Count)', fontsize=14)
plt.xlabel('Average Runtime (Minutes)', fontsize=12)
plt.ylabel('Average IMDb Score', fontsize=12)
plt.xlim(director_stats['avg_runtime'].min() * 0.95, director_stats['avg_runtime'].max() * 1.05)
plt.ylim(director_stats['avg_score'].min() * 0.95, director_stats['avg_score'].max() * 1.05)
plt.grid(True, linestyle='--', alpha=0.5)

# Create a manual legend for bubble size reference
for size in [50, 100, 150]:
    plt.scatter([], [], c='gray', alpha=0.6, s=np.sqrt(size) * 50, label=f'{size} Titles')
plt.legend(title='Approximate Title Count', scatterpoints=1, frameon=False, labelspacing=2, loc='upper left')

plt.tight_layout()
plt.savefig('director_performance_bubble_chart.png')
print("Saved bubble chart to director_performance_bubble_chart.png")
````

##### 1. Why did you pick the specific chart?

I have picked a Bubble Chart here which is an advanced form of a Scatter Plot, making it the perfect tool for multivariate analysis (visualizing three variables at once).

Here, we analyze the performance of the top 10 most frequent directors based on three metrics:

X-Axis: Average Runtime (Minutes)

Y-Axis: Average IMDb Score (Quality)

Bubble Size: Total Title Count (Productivity)

##### 2. What is/are the insight(s) found from the chart?

**Key Insights**
* The Most Valuable Directors (High Quality): Look at the bubbles positioned high on the chart (high Average IMDb Score). These are your prestige directors. They are reliable for high-quality content, regardless of the size of their bubble (productivity).
* The Volume Drivers (Low Quality): Look for directors with the largest bubbles (highest number of titles) positioned low on the chart. These directors are highly productive, but their typical title has a below-average score.

#### Chart - 12: **Subplot Chart (Multivariate Analysis through Biavariate Plots)**

````python
# Chart - 12 visualization code
import pandas as pd
import matplotlib.pyplot as plt

# Load the titles dataset
titles_df = pd.read_csv('titles.csv.zip')

# 1. Filter data to focus on the modern era (from 1950 onwards)
START_YEAR = 1950
trend_df = titles_df.dropna(subset=['release_year', 'type']).copy()
trend_df = trend_df[trend_df['release_year'] >= START_YEAR]

# 2. Group by release year and type to get two distinct time series
release_trend_by_type = trend_df.groupby(['release_year', 'type'])['id'].count().unstack(fill_value=0)

# Ensure both 'MOVIE' and 'SHOW' columns exist, even if one is empty in certain years
if 'MOVIE' not in release_trend_by_type.columns:
    release_trend_by_type['MOVIE'] = 0
if 'SHOW' not in release_trend_by_type.columns:
    release_trend_by_type['SHOW'] = 0

# 3. Create the Subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=True) # Share Y-axis for direct comparison of volume

# --- Subplot 1: Movie Releases ---
axes[0].plot(
    release_trend_by_type.index,
    release_trend_by_type['MOVIE'],
    color='darkorange',
    linewidth=2
)
axes[0].set_title('Annual Movie Releases Trend', fontsize=14)
axes[0].set_xlabel('Release Year', fontsize=10)
axes[0].set_ylabel('Number of Titles Released', fontsize=10)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[0].tick_params(axis='x', rotation=45)

# --- Subplot 2: Show Releases ---
axes[1].plot(
    release_trend_by_type.index,
    release_trend_by_type['SHOW'],
    color='darkviolet',
    linewidth=2
)
axes[1].set_title('Annual Show Releases Trend', fontsize=14)
axes[1].set_xlabel('Release Year', fontsize=10)
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust overall title and layout
fig.suptitle(f'Comparison of Content Release Trends ({START_YEAR} to Present)', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust for suptitle
plt.savefig('movie_show_release_subplots.png')
print("Saved subplots chart to movie_show_release_subplots.png")
````

#### Chart - 13 : **Violin plot with imdb score(numerical variable) accross age certification( categorical variable) using Bivariate Analysis**

````python
# Chart - 13 visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the titles dataset
titles_df = pd.read_csv('titles.csv.zip')

# 1. Data Wrangling: Identify Top 5 Age Certifications
# Fill missing age certifications temporarily for counting
titles_df['age_certification_clean'] = titles_df['age_certification'].fillna('Unrated/Missing')

# Get counts, drop the 'Unrated/Missing' category, and select the top 5
certification_counts = titles_df['age_certification_clean'].value_counts()
top_5_certifications = certification_counts.drop('Unrated/Missing', errors='ignore').head(5).index.tolist()

# 2. Filter data for plotting
# Filter for only the top 5 certifications and ensure IMDb score is present
plot_df = titles_df[
    titles_df['age_certification'].isin(top_5_certifications)
].dropna(subset=['imdb_score']).copy()

# Ensure the categories are ordered by median score for better comparison
median_order = plot_df.groupby('age_certification')['imdb_score'].median().sort_values(ascending=False).index

# 3. Create the Violin Plot
plt.figure(figsize=(8,5))

sns.violinplot(
    x='age_certification',
    y='imdb_score',
    data=plot_df,
    order=median_order,
    palette='viridis',
    inner='quartile', # Show the box plot elements inside the violin
    cut=0 # Ensures the violin doesn't extend beyond the data range
)

plt.title('Distribution of IMDb Scores by Top 5 Age Certifications', fontsize=16)
plt.xlabel('Age Certification', fontsize=12)
plt.ylabel('IMDb Score', fontsize=12)
plt.ylim(1, 10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('imdb_score_by_certification_violin_plot.png')
print("Saved violin plot to imdb_score_by_certification_violin_plot.png")
````

##### 1. Why did you pick the specific chart?

A Violin Plot is a visualization that goes beyond a standard box plot. It allows you to see the full probability density (the shape of the distribution) of a numerical variable across different categories.

I pick this chart to represent the comparison of  the IMDb Score Distribution across the top 5 most frequent Age Certifications in the content library.

The Violin Plot plots a Numerical Variable (IMDb Score) against a Categorical Variable (Age Certification), the plot analyzes the relationship between these two different data types, it is classified as a bivariate chart.

The width of the "violin" shape at any given score level represents how dense that score is for the corresponding certification.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the titles dataset
titles_df = pd.read_csv('titles.csv.zip')

# 1. Select the relevant numerical columns
numerical_cols = [
    'release_year',
    'runtime',
    'seasons',
    'imdb_score',
    'imdb_votes',
    'tmdb_popularity',
    'tmdb_score'
]
corr_df = titles_df[numerical_cols].copy()

# 2. Calculate the correlation matrix
correlation_matrix = corr_df.corr()

# 3. Create the Heatmap
plt.figure(figsize=(7,5))
sns.heatmap(
    correlation_matrix,
    annot=True,              # Show the correlation values on the heatmap
    cmap='coolwarm',         # Colormap from cool (negative) to warm (positive)
    fmt=".2f",               # Format annotations to two decimal places
    linewidths=.5,           # Draw lines between cells
    linecolor='black',
    cbar_kws={'label': 'Correlation Coefficient'}
)

plt.title('Correlation Heatmap of Key Numerical Content Metrics', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
print("Saved correlation heatmap to correlation_heatmap.png")
````

##### 1. Why did you pick the specific chart?

I pick this specific chart to represent a Correlation Heatmap which is a powerful multivariate visualization that uses color intensity to represent the strength and direction of the relationships between several numerical variables at once.

The chart uses a color scale (from dark red for strong negative correlation to dark blue for strong positive correlation) to show the linear relationship between every pair of metrics.

##### 2. What is/are the insight(s) found from the chart?

-- Key Insights from the above Correlation Heatmap :

* Strongest Positive Correlation (Blue Squares):

 IMDb Score vs. TMDB Score ($\mathbf{0.86}$): This is the strongest relationship, meaning that a high score on one platform almost always guarantees a high score on the other. This confirms that these two quality metrics are reliable proxies for each other.

 IMDb Votes vs. TMDB Popularity ($\mathbf{0.44}$): This positive link is intuitive: content with more user engagement (Votes) is also deemed more popular on TMDB.

* Weakest/No Correlation (White/Light Squares):

IMDb Score vs. Release Year ($\mathbf{-0.03}$): There is virtually no correlation between when content was released and its quality rating. Good content is released every year.

Runtime vs. IMDb Score ($\mathbf{0.05}$): The runtime of a title has almost no linear relationship with its quality. A title is not better or worse just because it is longer or shorter.

* Negative Correlation (Red Squares):

Seasons vs. Runtime ($\mathbf{-0.33}$): This is a significant finding. It means titles with more seasons tend to have a shorter runtime (per episode/unit). This is the expected trade-off in serialized content, where high episode count usually necessitates shorter individual episodes.

#### Chart - 15 - **Pair Plot (Multivariate Analysis)**

````python
# Pair Plot visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the titles dataset using the zip file path
titles_df = pd.read_csv("titles.csv.zip")

# 1. Select key numerical columns and clean data
numerical_cols = [
    'runtime',
    'imdb_score',
    'imdb_votes',
    'release_year'
]
pair_plot_df = titles_df[numerical_cols].dropna().copy()

# Filter out extreme outliers for better visualization in IMDb Votes and Runtime
# Filter Runtime: Focus on content under 200 minutes
pair_plot_df = pair_plot_df[pair_plot_df['runtime'] <= 200]
# Filter IMDb Votes: Exclude the top 1% of extreme outliers for clarity in scatter plots
pair_plot_df = pair_plot_df[pair_plot_df['imdb_votes'] <= pair_plot_df['imdb_votes'].quantile(0.99)]


# 2. Create the Pair Plot
sns.pairplot(
    pair_plot_df,
    plot_kws={'alpha': 0.3, 's': 10}, # Adjust point size and transparency for better visibility
    diag_kind='kde' # Use a Kernel Density Estimate instead of a simple histogram for smoother distribution view
)

# 3. Save the plot
plt.suptitle('Pairwise Relationships Among Key Content Metrics', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to prevent suptitle overlap
plt.savefig('pair_plot_key_metrics.png')
print("Saved pair plot to pair_plot_key_metrics.png")
````

##### 1. Why did you pick the specific chart?

I pick this chart to represent a Pair Plot (or scatter plot matrix) which is multivariate visualization that displays the pairwise relationships between multiple numerical variables in a dataset.

 It is widely used in exploratory data analysis (EDA) to quickly identify patterns, trends, correlations, and potential outliers among features.

##### 2. What is/are the insight(s) found from the chart?

From the above chart we got to know that the plot compares Runtime, IMDb Score, IMDb Votes (Popularity), and Release Year.

* Runtime vs. IMDb Score: You would see a big cloud of dots with no clear upward or downward slope.

* Release Year vs. IMDb Score: Again, you would see a flat distribution, content from the $1980$s or $1990$s holds up just as well as content released last year.

* IMDb Votes vs. IMDb Score:  For a title to be massively popular (high votes), it almost always has to be good (high score). The top-voted titles would be clustered in the high-score region.
* Library is Heavily Focused on the Last Decade : Release Year Distribution- The density plot (top-right of the plot) would show a huge spike for the $2010$s and $2020$s.
